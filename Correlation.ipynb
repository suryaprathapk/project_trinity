```python
import pandas as pd
```

```python
df_nfamp = pd.read_csv("./NFAMP1_data.csv")
df_nfamp
```

```python
NFAMP_column = df_nfamp.pop('NFAMP Price')
df_nfamp.insert(0, 'NFAMP Price', NFAMP_column)
df_nfamp.pop('NDC')
df_nfamp
```

```python
display(df_nfamp.corr().style.background_gradient(cmap='coolwarm'))
```

```python
df_nfamp = df_nfamp.loc[:, (df_nfamp != 0).any(axis=0)]
df_nfamp.pop('QYYYY')
df_nfamp
```

```python
display(df_nfamp.corr().style.background_gradient(cmap='coolwarm'))
```

```python
import matplotlib.pyplot as plt
from sklearn import linear_model

X = df_nfamp.values[:, 1:8]  # get input values from first two columns
y = df_nfamp.values[:, 0]  # get output values from last coulmn

model_ols =  linear_model.LinearRegression()
model_ols.fit(X,y) 
```

```python
coef = model_ols.coef_
intercept = model_ols.intercept_
print(list(df_nfamp.columns))
print('coef= ', coef)
print('intercept= ', intercept)
```

```python
features = list(df_nfamp.columns[1:])
features
```

```python
features_dict ={}
for i in range(len(features)):
    features_dict[features[i]] =coef[i]
features_dict
```

```python
predictedPrice = pd.DataFrame(model_ols.predict(X), columns=['Predicted Price']) # Create new dataframe of column'Predicted Price'
actualPrice = pd.DataFrame(y, columns=['Actual Price'])
actualPrice = actualPrice.reset_index(drop=True) # Drop the index so that we can concat it, to create new dataframe
df_actual_vs_predicted = pd.concat([actualPrice,predictedPrice],axis =1)
df_actual_vs_predicted.T
```

```python
plt.scatter(y, model_ols.predict(X))
plt.xlabel('Price From Dataset')
plt.ylabel('Price Predicted By Model')
plt.rcParams["figure.figsize"] = (10,6) # Custom figure size in inches
plt.title("Price From Dataset Vs Price Predicted By Model")
```
