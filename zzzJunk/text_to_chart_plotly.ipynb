```python
import plotly.graph_objects as go

class Node:
    def __init__(self, value):
        self.value = value
        self.children = []

def formula_to_tree(formula):
    stack = []
    for char in formula:
        if char == '(':
            stack.append(Node(char))
        elif char.isalpha():
            stack[-1].children.append(Node(char))
        elif char == ')':
            node = stack.pop()
            if stack:
                stack[-1].children.append(node)
            else:
                return node

def build_graph(root):
    if not root:
        return None
    if len(root.children) == 0:
        return go.Scatter(x=[0], y=[0], text=[root.value], mode="text")
    else:
        fig = go.Figure()
        x, y = [], []
        for i, child in enumerate(root.children):
            fig.add_trace(build_graph(child))
            x.append(i)
            y.append(0)
        fig.add_trace(go.Scatter(x=x, y=y, text=[root.value], mode="text"))
        return fig

formula = "((A+B)*(C-D))"
root = formula_to_tree(formula)
fig = build_graph(root)
fig.show()
```

```python
def formula_to_newick(formula):
    stack = []
    for char in formula:
        if char == '(':
            stack.append(char)
        elif char.isalpha():
            stack.append(char)
        elif char == ')':
            node = stack.pop()
            while node != '(':
                if node.isalpha():
                    if len(stack) > 0:
                        if stack[-1].isalpha():
                            stack[-1] = stack[-1]+node
                        else:
                            stack[-1] = stack[-1]+','+node
                    else:
                        stack.append(node)
                else:
                    stack.append(node)
                if len(stack)>0:
                    node = stack.pop()
                else:
                    break
    return ''.join(stack)

formula = "((A+B)*(C-D))"
newick_formula = formula_to_newick(formula)
print(newick_formula)
```

```python
import networkx as nx
import plotly.graph_objects as go

def formula_to_tree(formula):
    formula = formula.replace(" ","")
    formula = formula.split("=")[1]
    stack = []
    for char in formula:
        if char in "ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz":
            stack.append(char)
        elif char in "+":
            right = stack.pop()
            left = stack.pop()
            stack.append(Node(char,[left,right]))
    return stack[0]

class Node:
    def __init__(self, value, children=[]):
        self.value = value
        self.children = children

def build_graph(root):
    G = nx.DiGraph()
    build_graph_recursive(G, root)
    return G

def build_graph_recursive(G, root):
    if not root:
        return
    G.add_node(root.value)
    for child in root.children:
        G.add_edge(root.value, child)
        build_graph_recursive(G, child)

formula = "y = a+b+c+d"
root = formula_to_tree(formula)
G = build_graph(root)

pos = nx.nx_agraph.graphviz_layout(G, prog='dot')
edge_x = []
edge_y = []
for edge in G.edges():
    x0, y0 = pos[edge[0]]
    x1, y1 = pos[edge[1]]
    edge_x.append(x0)
    edge_x.append(x1)
    edge_x.append(None)
    edge_y.append(y0)
    edge_y.append(y1)
    edge_y.append(None)

edge_trace = go.Scatter(
    x=edge_x, y=edge_y,
    line=dict(width=0.5, color='#888'))
```
